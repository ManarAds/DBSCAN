```python
import pandas as pd
import matplotlib.pyplot as plt
wine_date=pd.read_csv('wine.csv')
print(wine_date)
color=plt.cm.rainbow(np.linspace(0,1,len(wine_date)))
plt.scatter(wine_date.values[: , 0] , wine_date.values[:,1] , c =color)
plt.xlabel("OD_read")
plt.ylabel("Proline")
plt.show()
```

```python
#DBSCAN

paramter=[1.1 , 2] , [1.1 , 3] , [1.1,4] , [ 2.1,4],[4.1,5] , [4.1 , 4]
for eps , min in paramter :
 db=DBSCAN(eps= eps, min_samples=min , metric="manhattan")
 plt.scatter(wine_date.values[: , 0] , wine_date.values[:,1], c=db.fit_predict(wine_date))
 plt.title("DBSCAN for eps=%s and minPts=%s" % (eps, min))
 plt.show()
```

```python
from sklearn.cluster import KMeans , AgglomerativeClustering , DBSCAN
from sklearn.metrics import silhouette_score
km=KMeans(4)
km.fit(wine_date)
km_lable=km.fit_predict(wine_date)
print(f'k-means is {silhouette_score(wine_date , km_lable)}')

ac=AgglomerativeClustering( 3, linkage='average')
ac_lable=ac.fit_predict(wine_date)
print(f'aggumlative is {silhouette_score(wine_date , ac_lable)}')

```

```python
db_param_options = [[20,5],[25,5],[30,5],[25,7],[35,7],[35,3],[36,5],[55,4]]

for eps , min in db_param_options:
    Db=DBSCAN(eps=eps , min_samples=min )
    db_cluster=Db.fit_predict(wine_date)
    print(f'the eps is :{eps} min_sample is : {min}  ')
    print(f"DBSCAN clustring is {silhouette_score(wine_date,db_cluster)} ")
```

```python

plt.scatter(wine_date.values[: , 0] , wine_date.values[:,1] , c=km_lable , s=50 )
plt.title("Wine Clusters from K-Means")
plt.show()


plt.scatter(wine_date.values[: , 0] , wine_date.values[:,1] , c=ac_lable  , s=50)
plt.title("Wine Clusters from Agglomrative clustring ")
plt.show()

plt.scatter(wine_date.values[: , 0] , wine_date.values[:,1] , c=db_cluster  ,s=50)
plt.title("Wine Clusters from DBSCAN")

plt.show()
```

```python
print("Silhouette Scores for Wine Dataset:\n")
print("K-Means Clustering: ", silhouette_score(wine_date, km_lable))
print("Agg Clustering: ", silhouette_score(wine_date, ac_lable))
print("DBSCAN Clustering: ", silhouette_score(wine_date, db_cluster))
```
